```python
def create_id(
    path, file_name, lname, fname, nation, id_num, exp_date, birth_date, gender
):
    id_template = cv2.imread("./documents/personal_id.jpeg")
    person = cv2.imread(path + file_name)
    person = cv2.cvtColor(person, cv2.COLOR_BGR2RGB)
    id_template = cv2.cvtColor(id_template, cv2.COLOR_BGR2RGB)

    id_template[
        200 : 200 + cv2.resize(person, (195, 265)).shape[0],
        70 : 70 + cv2.resize(person, (195, 265)).shape[1],
    ] = cv2.resize(person, (195, 265))
    id_template[
        290 : 290 + cv2.resize(person, (95, 125)).shape[0],
        625 : 625 + cv2.resize(person, (95, 125)).shape[1],
    ] = cv2.resize(person, (95, 125))
    font = cv2.FONT_HERSHEY_COMPLEX
    cv2.putText(
        id_template, "Watermark", (230, 500), font, 3, (195, 105, 255), 2, cv2.LINE_AA
    )
    cv2.putText(id_template, lname, (300, 160), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, fname, (300, 225), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, nation, (300, 275), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, id_num, (300, 345), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, exp_date, (300, 395), font, 1, (0, 0, 0), 2, cv2.LINE_AA)

    cv2.putText(id_template, birth_date, (550, 275), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, gender, (550, 345), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(id_template, "", (620, 470), font, 1, (0, 0, 0), 2, cv2.LINE_AA)
    return id_template
```

```python
def create_license(
    path,
    file_name,
    lname,
    fname,
    place_birth,
    date,
    begin,
    end,
    who,
    pesel,
    gender,
    sign,
    type,
    number,
):
    passport = cv2.imread("./documents/passport.jpeg")
    passport = cv2.cvtColor(passport, cv2.COLOR_BGR2RGB)
    person = cv2.imread(path + file_name)
    person = cv2.cvtColor(person, cv2.COLOR_BGR2RGB)

    passport[
        200 : 200 + cv2.resize(person, (250, 355)).shape[0],
        40 : 40 + cv2.resize(person, (250, 355)).shape[1],
    ] = cv2.resize(person, (250, 355))
    font = cv2.FONT_HERSHEY_COMPLEX

    cv2.putText(passport, lname, (350, 190), font, 0.9, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, fname, (350, 240), font, 0.9, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, date, (350, 300), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, place_birth, (650, 300), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, begin, (350, 410), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, end, (350, 465), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, who, (350, 520), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, pesel, (850, 350), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, gender, (850, 410), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, type, (350, 125), font, 0.7, ( 0,  0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, number, (650, 125), font, 0.7, ( 0,  0, 0), 2, cv2.LINE_AA)
    cv2.putText(passport, sign, (850, 520), font, 0.7, (0, 0, 0), 2, cv2.LINE_AA)

    return passport
```
